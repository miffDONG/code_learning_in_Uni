## 자주 사용하지 않는 단어 제외하기
### freqdist로 빈도분포 만들기
- Text_string column의 요소를 ' '로 이어붙인 뒤, 해당 text를 word_tokenize를 이용하여 tokenize 진행
- tokenize된 list를 nltk의 FreqDist를 활용하여 단어의 빈도수 계산

```python
# Text_String column은 Text_token에서 한 글자 단어를 제외하고 문장으로 만든 column입니다.
# Text_string_token은 Text_string에 대해 tokenize를 적용한 column입니다.
# 해당 셀을 실행한 뒤 코드를 작성하세요.
train['Text_string'] = train['Text_token'].apply(lambda x: ' '.join([item for item in x if len(item)>2]))
train = train.drop(['Text_token'], axis = 1)
train['Text_string_token']=train['Text_string'].apply(nltk.tokenize.word_tokenize)
train.head()
```

```python
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize

# Write your codes here
all_tokens = [token for sublist in train['Text_string_token'] for token in sublist]

fdist = FreqDist(all_tokens)

fdist
```

### 빈도수가 낮은 단어 제거
- 위 문제에서 구한 FreqDist dictionary를 이용하여 Text_string_token column에 대해 빈도수가 2 이상인 단어만으로 구성 후 Text_string_fdist column을 만드시오.

## Encoding & Padding
- train dataframe의 Text_string_fdist를 encoding & padding 진행하기.

### Encoding
- Text_string_fdist와 fdist를 이용하여 text를 encoding 하시오.
- Text_string_fdist의 각 요소별 단어를 fdist의 value로 mapping을 하는 작업입니다.

### Padding
- encoding된 vector의 요소 중 가장 긴 문자열의 길이를 기준으로 padding을 진행하시오.(padding number = 0)
- encoding 된 vector의 요소 중 가장 긴 리스트의 길이보다 짧은 리스트는 제일 긴 리스트의 길이와의 차이만큼 0으로 채우는 작업입니다.

```python
# Write your codes here

encoded = train['Text_string_fdist'].apply(lambda x:[fdist[word] for word in x])

max_len = max(len(enc) for enc in encoded)

encoded = [lst + [0] * (max_len - len(lst)) for lst in encoded]
encoded[0][:30]# encoding&padding 된 list의 일부 출력.
```

## Train / Test split

- X_data : encoded를 nparray로 변환한 다음 float32로 자료형 변환
- y_data = train dataframe의 Score를 numpy로 변환한 다음 float32로 자료형 변환
- train_test_split을 이용하여 train_valid로 dataset 분할 : test_size = 0.2, random_state = 0, stratify = y_data

```python
from sklearn.model_selection import train_test_split

X_data = np.array(encoded, dtype=np.float32)
y_data = train['Score'].values.astype(np.float32)

X_train, X_valid, y_train, y_valid = train_test_split(X_data,y_data,test_size=0.2,random_state=0,stratify=y_data)
```

## Train 중 validation을 진행하는 이유를 서술하시오.

- validation '검증' 단계의 이유를 알기 전에, 먼저 우리는 모델을 만드는 이유를 알 필요가 있습니다. 모델은 예측을 위한 존재로, Unseen data에 대한 예측 또는 분류를 목적으로 하며 범용적으로 사용할 수 있는 모델을 만들고자 합니다.
- validation set은 test set과 동일하게 학습에 직접적으로 관여하지 않고, 모델 성능을 검증해준다는
 공통점을 가지지만 fine tuning에 사용된다는 점에서 차이가 있습니다. 모델 평가로 넘어가기 전에 'epoch,lr' 등 하이퍼파라미터 조정을 통해 성능을 올리는데 도움을 줍니다.

## Model

- 간단한 linear regression model 구현해보기.

```python
class linearRegression(torch.nn.Module):
    def __init__(self, inputSize, outputSize):
        super(linearRegression, self).__init__()
        self.linear = torch.nn.Linear(inputSize,outputSize)
    

    def forward(self, x):
        out = self.linear(x)

        return out
```

```python
inputDim = X_train.shape[1]
outputDim = 1

learningRate = 1e-6
epochs = 2000

model = linearRegression(inputDim, outputDim)

if torch.cuda.is_available():
    model.cuda()
```

## Loss function, optimizer 정의하기
- loss_fn : pytorch에 구현되어있는 MSE loss 사용하기
- optimizer : pytorch에 구현되어 있는 SGD 사용하기 - 주어진 learning rate 사용

```python
from torch import nn
import torch.optim as optim

loss_fn = nn.MSELoss()

optimizer = optim.SGD(model.parameters(), lr=learningRate)
```

## 코드의 빈칸 채워넣기
- 코드 중 _____로 적혀진 부분 채워넣기
- loss의 값이 일치하지 않을 수 있습니다.

```python
from torch.autograd import Variable

train_loss, train_epoch = [], []
random_seed = 42

if torch.cuda.is_available() :
    torch.cuda.manual_seed_all(random_seed)
else:
    torch.manual_seed(random_seed)

for epoch in range(epochs):
    if torch.cuda.is_available():
        inputs = Variable(torch.from_numpy(X_train).cuda())
        labels = Variable(torch.from_numpy(y_train).cuda())
    else:
        inputs = Variable(torch.from_numpy(X_train))
        labels = Variable(torch.from_numpy(y_train))
    
    optimizer.zero_grad() # .grad를 0으로 초기화

    outputs = model(inputs)
    
    loss =  loss_fn(outputs,labels)

    loss.backward() 
    
    optimizer.step() # parameter update(최적화 단계)

    if epoch % 100 == 0:
        print('epoch {}, loss {}'.format(epoch, loss.item()))
        train_loss.append(loss.item())
        train_epoch.append(epoch)

    if epoch % 200 == 0 :
        with torch.no_grad():
            valid_output = model(torch.from_numpy(X_valid))
            valid_loss = loss_fn(valid_output, torch.from_numpy(y_valid))
            print('epoch {}, valid_loss {}'.format(epoch, valid_loss.item()))
```

## Learning rate에 따른 loss 비교
- 아래의 코드의 빈칸을 채우고 실행한 뒤, training 중 loss의 값이 inf가 되는 이유를 learning rate과 관련하여 간단히 서술하시오
- 아래의 코드에서 변수명은 model이 아닌 models입니다. 주의하세요.

```python
inputDim = X_train.shape[1]
outputDim = 1

learningRate = 1e-5
epochs = 200

models = linearRegression(inputDim, outputDim)

if torch.cuda.is_available():
    models.cuda()

loss_fn = torch.nn.MSELoss() # MSE loss function
optimizer = optim.SGD(models.parameters(),lr=learningRate) # SGD optimizer → 주어진 learning_rate와 models 사용

random_seed = 42
if torch.cuda.is_available() :
    torch.cuda.manual_seed_all(random_seed)
else:
    torch.manual_seed(random_seed)

for epoch in range(epochs):

    if torch.cuda.is_available():
        inputs = Variable(torch.from_numpy(X_train).cuda())
        labels = Variable(torch.from_numpy(y_train).cuda())
    else:
        inputs = Variable(torch.from_numpy(X_train))
        labels = Variable(torch.from_numpy(y_train))
    
    optimizer.zero_grad()

    outputs = models(inputs)

    loss = loss_fn(outputs,labels)
    loss.backward()

    optimizer.step()

    if epoch % 100 == 0:
        print('epoch {}, loss {}'.format(epoch, loss.item()))
```
