```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.__version__
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.histplot(data=tips,x='tip')
```

```python
sns.barplot(data=tips,x='day',y='tip',errorbar=None,hue= 'time')
```

```python
sns.boxplot(data=tips, x='day', y='tip', hue='time')
```

```python
sns.violinplot(data=tips,x='day',y='tip',hue='smoker',split=True)
```

```python
fig, axes = plt.subplots(1,2)

sns.boxplot(data=tips,x='day',y='tip',hue='smoker',ax=axes[0])
sns.violinplot(data=tips,x='day',y='tip',hue='smoker',split=True,ax=axes[1])
plt.show()
```
