```python
from langchain.embeddings import AzureOpenAIEmbeddings

embeddings_model = AzureOpenAIEmbeddings(
    deployment="dev-text-embedding-ada-002",
    chunk_size=500
)
```

```python
embedding_examples = {
    '안녕하세요',
    '제 이름은 홍길동 입니다.',
    '이름은 무엇입니까?',
    '랭체인은 유용합니다.',
    'Hello World'
}
```

```python
embeddings = embeddings_model.embed_documents(embedding_examples)

print('Total embeddings:' + str(len(embeddings)))
print('Dimension of each embedding:' + str(len(embeddings[0])))
```

```python
q = '이 대화에서 언급된 이름은 무엇입니까?'
a = '이 대화에서 언급된 이름은 홍길동 입니다.'

embeddings_q = embeddings_model.embed_query(q)
embeddings_a = embeddings_model.embed_query(a)
```

```python
from numpy import dot
from numpy.linalg import norm
import numpy as np

def cos_sim(a, b):
    return dot(a, b) / (norm(a) * norm(b))
```

```python
print(cos_sim(embeddings_q, embeddings_a))
print(cos_sim(embeddings_q, embeddings[1]))
print(cos_sim(embeddings_q, embeddings[3]))
```

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings

model_name = "BAAI/bge-small-en"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': True}
hf = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
embedding_examples = {
    'today is monday',
    'weather is nice today',
    'what is the problem',
    'langchain is useful',
    'Hello World',
    'my name is morris'
}
```

```python
embeddings = hf.embed_documents(embedding_examples)

print('Total embeddings:' + str(len(embeddings)))
print('Dimension of each embedding:' + str(len(embeddings[0])))
```

```python
q = 'Hello? who is this?'
a = 'hi this is harrison'

embeddings_q = hf.embed_query(q)
embeddings_a = hf.embed_query(a)

print(cos_sim(embeddings_q, embeddings_a))
print(cos_sim(embeddings_q, embeddings[1]))
print(cos_sim(embeddings_q, embeddings[5]))
```

```python
embedding_examples = {
    '안녕하세요',
    '제 이름은 홍길동 입니다.',
    '이름은 무엇입니까?',
    '랭체인은 유용합니다.',
    '홍길동 아버지의 이름은 홍상직 입니다. '
}

embeddings = hf.embed_documents(embedding_examples)

q = '홍길동은 아버지를 아버지라 부르지 못했습니다. 홍길동의 아버지의 이름은 무엇입니까?'
a = '홍길동 아버지는 엄했습니다.'

embeddings_q = hf.embed_query(q)
embeddings_a = hf.embed_query(a)

print(cos_sim(embeddings_q, embeddings_a))
print(cos_sim(embeddings_q, embeddings[1]))
print(cos_sim(embeddings_q, embeddings[3]))
print(cos_sim(embeddings_q, embeddings[4]))
```
