# Matplot lib

matlab vs 객체(like python)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.__version__
```

```python
x = np.arange(0,50,0.1)
y = np.sin(x)
z = np.cos(x)

def sinShow(x,y,z):
    plt.figure(figsize=(13,5))
    plt.plot(x,y,color='green',label='sin')
    plt.plot(x,z,color='red',label='cos')
    plt.title("line chart")
    plt.xlabel("X values")
    plt.ylabel("y values")
    plt.legend()
    plt.show()

sinShow(x,y,z)
```

```python
def objectiveShow(x,y,z):
    fig, ax = plt.subplots()
    fig.set_size_inches(13,5)
    ax.plot(x,y,label='sin(x)')
    ax.plot(x,z,label='cos(x)')
    ax.set_xlabel('X values')
    ax.set_ylabel('y values')
    ax.legend()
    
objectiveShow(x,y,z)
```

## Boxplot

이상치 확인 : 이상할 가능성이 있음을. 많이 처리한다고 좋은건 아님.

```python
x = np.random.normal(0,2.0,1000)
y = np.random.normal(-3.0,1.5,500)
z = np.random.normal(1.2,1.5,1500)

plt.boxplot([x,y,z])
plt.show()
```
