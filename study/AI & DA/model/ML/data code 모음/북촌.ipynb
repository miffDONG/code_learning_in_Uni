```python
import os
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_1=pd.read_csv('C://Users/hyundong/Desktop/buk/202109-12_BUKCHON_CCTV.csv',encoding='CP949')
data_2=pd.read_csv('C://Users/hyundong/Desktop/buk/202112-03_BUKCHON_CCTV.csv',encoding='CP949')
data_3=pd.read_csv('C://Users/hyundong/Desktop/buk/202203-06_BUKCHON_CCTV.csv',encoding='CP949')
data_4=pd.read_csv('C://Users/hyundong/Desktop/buk/202206-09_BUKCHON_CCTV.csv',encoding='CP949')
data_5=pd.read_csv('C://Users/hyundong/Desktop/buk/202209-12_BUKCHON_CCTV.csv',encoding='CP949')
```

```python
fi_data = pd.concat([data_1,data_2,data_3,data_4,data_5])
fi_data = fi_data.sort_values(by=['주소','측정 종료 시간'])
fi_data.columns = ['DEVICEID','DEVICENAME','DESCRIPTION','STARTTIME','ENDTIME','INCOUNT','OUTCOUNT','HOUR']
fi_data = fi_data.reset_index()
fi_data
```

```python

month=[]
hour=[]
for i in fi_data.loc[:,'ENDTIME']:
    _date = i.split(' ')[0]
    _hour = i.split(' ')[1]
    _date_items = _date.split('-')
    _hour_items = _hour.split(':')
    _month = _date_items[1]
    month.append(_month)
    _hour = _hour_items[0]
    hour.append(_hour)
    
       
fi_data['MONTH'] = month
fi_data['HOUR'] = hour 

fi_data
```

```python
# fi_data['TOTAL'] = fi_data['INCOUNT'] + fi_data['OUTCOUNT']
fi_data['INFLOW'] = fi_data['INCOUNT'] - fi_data['OUTCOUNT']
fi_data
# date_time = fi_data.group_by(['DESCRIPTION'],['DATE']).mean()
```

```python
plt.figure(figsize=(15,5))
fi_data.groupby(['HOUR'])['INCOUNT'].mean().plot(label='INFLOW')
fi_data.groupby(['HOUR'])['OUTCOUNT'].mean().plot(label='OUTFLOW')
plt.title('FLOATING POPULATION')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
fi_data.groupby(['MONTH'])['INCOUNT'].mean().plot(label='INFLOW')
fi_data.groupby(['MONTH'])['OUTCOUNT'].mean().plot(label='OUTFLOW')
fi_data.groupby(['MONTH'])['INFLOW'].mean().plot(label='NETINFLOW')
plt.title('FLOATING POPULATION')
plt.legend()
plt.show()
```
