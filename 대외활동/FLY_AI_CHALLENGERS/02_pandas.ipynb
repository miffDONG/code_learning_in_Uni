## loc vs iloc

- loc : 조건문을 통한 값 반환
- iloc : index를 통한 값 반환

```python
game_data.loc[(game_data.global_sales>30),'jp_sales'] # Series
game_data.loc[(game_data.global_sales>30),['jp_sales']] # Dataframe
```

```python
game_data.iloc[0:4,[1,3]] # iloc은 조건문을 못씀. 
```
