```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ad_10000records.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head(15)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
# 하루에 사이트에 얼마나 시간을 보내는지에 따른 광고 클릭 여부
sns.boxenplot(x = 'clicked_on_ad', y = 'daily_time_spent_on_site',data = df)
```

```python
# 나이대별 광고 클릭 여부 
sns.boxenplot(x = 'clicked_on_ad', y = 'age', hue='gender', data = df)
```

```python
# area_income에 따른 광고 클릭 여부
sns.boxenplot(x = 'clicked_on_ad', y = 'area_income',hue = 'gender', data = df)
```

```python
# 하루에 인터넷 사용량에 따른 광고 클릭 여부
sns.boxenplot(x = 'clicked_on_ad', y = 'daily_internet_usage',data = df)
```

```python
# 성별에 따른 광고 클릭 여부
sns.barplot(x = 'clicked_on_ad', y = 'gender', data = df)
```

```python
#area income 의 분포
sns.distplot(df['area_income'])

```

```python
#area_income 10000씩 나누어서 그룹화
df['area_income_group'] = (df['area_income'] // 10000)
```

```python
# area_income_group 별로 클릭한 비율
sns.barplot(x = 'area_income_group', y = 'clicked_on_ad', data = df)
```

```python
# area_income_group 과 gender 별로 클릭한 비율
sns.barplot(x = 'area_income_group', y = 'clicked_on_ad', hue = 'gender', data = df)
```

```python
# area_income 그룹별과 Daily Time Spent on Site

sns.boxenplot(x = 'area_income_group', y = 'daily_time_spent_on_site', hue = 'gender' , data = df)
```

```python
sns.boxenplot(x = 'clicked_on_ad', y = 'daily_time_spent_on_site', hue = 'area_income_group', data = df)
```

```python
df[(df['area_income_group'] == 4) & (df['gender'] == 'Female')]['ad_topic_line'].value_counts().head(20).plot(kind='barh')
```

```python
# 나이별로 그룹화 
df['age_group'] = (df['age'] // 10 * 10)
```

```python
#나이 그룹별 클릭한 비율
sns.barplot(x = 'age_group', y = 'clicked_on_ad', hue = 'gender',  data = df)
```

```python
# 40 50 60의 주요 클릭 광고 주제
df[(df['age_group'] >= 40) & (df['clicked_on_ad'] == 1)]['ad_topic_line'].value_counts().head(20).plot(kind='barh')
```

```python
# income 그룹별 나이별 클릭한 비율
sns.heatmap(df.groupby(['area_income_group', 'age_group'])['clicked_on_ad'].mean().unstack(), annot=True, fmt=".2f")
```

```python
# income 그룹별 나이별 클릭한 비율
sns.heatmap(df.groupby(['area_income_group', 'age_group'])['daily_time_spent_on_site'].mean().unstack(), annot=True, fmt=".2f")
```

```python
# income 그룹별 나이별 클릭한 비율
sns.heatmap(df.groupby(['area_income_group', 'age_group'])['daily_internet_usage'].mean().unstack(), annot=True, fmt=".2f")
```
